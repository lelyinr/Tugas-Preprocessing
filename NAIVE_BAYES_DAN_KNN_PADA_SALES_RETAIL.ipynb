MEMISAHKAN VARIABEL X DAN Y

Dalam analisis klasifikasi menggunakan dataset retail, pemisahan data menjadi **data X** dan **data y** merupakan langkah awal yang sangat penting dalam proses persiapan data. Data X terdiri dari fitur-fitur independen seperti *Age*, *Quantity*, *Price per Unit*, dan *Total Amount* yang berfungsi sebagai input bagi model untuk mempelajari pola-pola dalam data. Sementara itu, data y berisi variabel target atau label, yaitu kolom *Product Category*, yang ingin diprediksi oleh model. Pemisahan ini memungkinkan model untuk belajar dari fitur-fitur dalam data X tanpa terlebih dahulu mengetahui labelnya, lalu diuji menggunakan data y untuk mengevaluasi seberapa akurat prediksinya. Proses ini tidak hanya membantu dalam pelatihan model, tetapi juga sangat penting dalam validasi dan pengukuran performa model klasifikasi terhadap data baru.

```python
#memisahkan data menjadi variabel x dan y
X=retail.drop(['Product Category'], axis=1) # Changed 'product category' to 'Product Category'
y=retail['Product Category'] # Changed 'product category' to 'Product Category'
```

MEMBAGI DATA MENJADI DATA TRAINING DAN TESTING

Data training merupakan bagian dari dataset yang digunakan untuk *melatih* model dalam proses machine learning. Melalui data ini, model belajar mengenali pola dan hubungan antara fitur-fitur (variabel independen) dengan target yang ingin diprediksi (variabel dependen). Selama pelatihan, model menyesuaikan parameter-parameter internalnya agar dapat membuat prediksi yang tepat berdasarkan pola yang teridentifikasi dari data training tersebut.

Sebaliknya, data testing berfungsi sebagai alat evaluasi. Bagian dataset ini *tidak* disertakan dalam proses pelatihan, melainkan digunakan untuk mengukur sejauh mana model mampu menerapkan pola yang telah dipelajarinya ke data baru yang belum pernah dilihat. Dengan menggunakan data testing, kita dapat menilai kemampuan model dalam melakukan generalisasi, bukan sekadar mengingat data pelatihan, sehingga dapat menunjukkan performa nyatanya dalam situasi dunia nyata.

Dalam hal proporsi, pembagian antara data training dan testing adalah aspek penting dalam pembangunan model yang handal. Meskipun tidak ada standar mutlak, pendekatan umum yang sering digunakan adalah pembagian 70:30 atau 80:20, di mana mayoritas data digunakan untuk pelatihan dan sisanya untuk pengujian. Untuk dataset berukuran besar, pembagian bisa lebih condong ke pelatihan, seperti 90:10 atau bahkan 95:5, asalkan tetap menyisakan cukup data uji untuk evaluasi model yang akurat dan objektif.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Fungsi `train_test_split` dari pustaka Scikit-Learn digunakan untuk memisahkan dataset menjadi dua bagian utama, yaitu data training dan data testing. Dalam penerapannya, baris kode `X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)` akan membagi data fitur (X) dan target (y) ke dalam empat bagian: X_train dan y_train untuk melatih model, serta X_test dan y_test untuk menguji kinerjanya. Parameter `test_size=0.2` menunjukkan bahwa 20% dari data akan digunakan untuk pengujian, sedangkan sisanya untuk pelatihan. Sementara itu, `random_state=0` menjamin bahwa pembagian data bersifat konsisten setiap kali kode dijalankan. Dengan pemisahan ini, kita dapat menilai seberapa baik model dalam mengenali pola dari data pelatihan dan mengaplikasikannya secara akurat pada data baru yang belum pernah dilihat sebelumnya.

```python
from sklearn.preprocessing import MinMaxScaler

# Inisialisasi scaler
scaler = MinMaxScaler()

# Select only numeric features for scaling
# Correct the column name from 'Tax 5%' to the actual column name in your DataFrame
numeric_features = ['Age', 'Quantity', 'Price per Unit', 'Total Amount',]

# Fit and transform on numeric features only
X_train[numeric_features] = scaler.fit_transform(X_train[numeric_features])
X_test[numeric_features] = scaler.transform(X_test[numeric_features])

# Print the transformed data
print(X_train)
```

`MinMaxScaler` adalah salah satu teknik normalisasi yang disediakan oleh Scikit-Learn, yang berfungsi untuk mengubah skala nilai fitur ke dalam rentang tertentu, umumnya antara 0 hingga 1. Pada implementasinya, fungsi `MinMaxScaler().fit_transform(X_train)` diterapkan untuk menormalkan data pelatihan (`X_train`), sementara `MinMaxScaler().fit_transform(X_test)` digunakan untuk menormalkan data pengujian (`X_test`). Proses normalisasi ini bertujuan agar semua fitur berada dalam skala yang seragam, sehingga model dapat bekerja lebih efisien dan menghasilkan prediksi yang lebih stabil. Metode `fit_transform()` sendiri menjalankan dua proses sekaligus: pertama, `fit()` menghitung nilai minimum dan maksimum dari setiap fitur dalam data pelatihan; kemudian `transform()` menggunakan nilai tersebut untuk mengubah setiap fitur menjadi skala yang sesuai dengan rentang yang ditentukan. Normalisasi semacam ini sangat penting, terutama saat fitur memiliki skala yang berbeda-beda, agar tidak mendominasi proses pembelajaran model.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import MinMaxScaler

# Assuming 'retail' is your DataFrame
# ... (Your previous code to load and prepare 'retail' DataFrame)

# Convert 'Date' column to datetime objects if it's not already
if 'Date' in retail.columns:
    retail['Date'] = pd.to_datetime(retail['Date'])

# Extract numeric features for scaling, excluding 'Date' and target variable
numeric_features = ['Age', 'Quantity', 'Price per Unit', 'Total Amount']

# Separate features (X) and target (y)
X = retail[numeric_features]  # Only include numeric features
y = retail['Product Category']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Initialize scaler and fit_transform on training data
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)  # Fit and transform on X_train
X_test = scaler.transform(X_test)  # Transform X_test using the same scaler

# Initialize and fit the Gaussian Naive Bayes model
nb = GaussianNB()
nb.fit(X_train, y_train)

# ... (Rest of your code)
```

```python
# compute the accuracy of train set
nb.score(X_train, y_train)
```

```python
classif_name = ['NB']
y_predict = {}
y_predict['NB'] = nb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

for name in classif_name:
    print('{0} accuracy = {1:.4f}'.format(name, accuracy_score(y_test, y_predict[name])))

```

```python
#Menguji parameter yang paling optimum untuk model

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV

from sklearn.model_selection import StratifiedKFold

cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=100)
# Tentukan parameter yang ingin diuji
parameters = {
    'hidden_layer_sizes': [(50,), (100,), (50, 30), (100, 50)],
    'learning_rate_init': [0.001, 0.01, 0.1],
    'max_iter': [100, 200, 300]
}

# Buat model MLPClassifier
clf = MLPClassifier(random_state=100)

# Inisialisasi GridSearchCV
grid_search = GridSearchCV(clf, parameters, cv=cv_strategy)
# cv merupakan jumlah lipatan validasi silang (cross-validation)

# Lakukan pencarian grid pada data training
grid_search.fit(X, y)
```

Tentukan Parameter yang Akan Diuji: Parameter-parameter yang ingin diuji didefinisikan dalam bentuk kamus. Dalam contoh ini, parameter yang diuji mencakup `hidden_layer_sizes` (berbagai kombinasi jumlah dan ukuran lapisan tersembunyi), `learning_rate_init` (berbagai tingkat pembelajaran awal), dan `max_iter` (berbagai jumlah iterasi).

Inisialisasi Model: Model Neural Network Multi-Layer Perceptron (MLP) dibuat menggunakan MLPClassifier(), yang dipasang dengan parameter default.

Agar hasil dari GridSearchCV konsisten di setiap percakapan, kamu bisa menambahkan argumen `random_state` pada GridSearchCV dan MLPClassifier. Penggunaan `random_state` akan memastikan bahwa keputusan yang melibatkan unsur acak (seperti pembagian data untuk validasi silang atau inisialisasi parameter awal pada Neural Network) dilakukan dengan cara yang sama setiap kali percobaan dilakukan. Perlu dicatat bahwa nilai `random_state` tidak perlu selalu 100, angka lainnya juga dapat digunakan, selama nilai yang sama dipertahankan dalam setiap percobaan untuk mendapatkan hasil yang konsisten.

Inisialisasi GridSearchCV: GridSearchCV digunakan untuk melakukan pencarian parameter terbaik dengan bantuan cross-validation. Argumen `cv=5` menunjukkan bahwa validasi silang dilakukan dengan 5 lipatan (5-fold cross-validation).

Pencarian Grid pada Data Training: Perintah `grid_search.fit(X, y)` digunakan untuk menjalankan pencarian parameter terbaik dengan memanfaatkan data pelatihan (X adalah fitur, y adalah target atau label).

GridSearchCV akan mengevaluasi semua kombinasi parameter yang telah ditentukan dan melatih serta mengevaluasi model pada setiap kombinasi tersebut. Tujuan utamanya adalah untuk menemukan kombinasi parameter yang memberikan performa terbaik berdasarkan metrik evaluasi yang telah ditentukan (seperti akurasi, presisi, atau recall).

Setelah proses pencarian selesai, parameter terbaik dapat diakses melalui atribut `best_params_` pada objek grid_search. Atribut ini akan menampilkan kombinasi parameter terbaik yang ditemukan selama pencarian grid.

```python
# Tampilkan parameter terbaik
print("Parameter terbaik:", grid_search.best_params_)
```

Dalam contoh ini, model MLPClassifier dibuat dengan sejumlah parameter yang ditetapkan secara manual. Parameter-parameter tersebut meliputi:

learning_rate_init=0.1: Parameter ini menentukan tingkat pembelajaran awal, yang mengontrol seberapa besar model menyesuaikan bobotnya berdasarkan kesalahan pada setiap iterasi pelatihan.

max_iter=100: Parameter ini menunjukkan jumlah maksimum iterasi yang akan dilakukan selama pelatihan model, yang mengatur berapa banyak epoch yang akan diproses dalam sesi pelatihan.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Inisialisasi model KNN
knn_model = KNeighborsClassifier(n_neighbors=5)  # kamu bisa coba nilai k lain seperti 3, 7, dst.

# Melatih model KNN
knn_model.fit(X_train, y_train)

# Melakukan prediksi pada data uji
y_pred_knn = knn_model.predict(X_test)

# Evaluasi hasil prediksi
print("Accuracy Score:", accuracy_score(y_test, y_pred_knn))
print("\nClassification Report:\n", classification_report(y_test, y_pred_knn))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
```

Berdasarkan hasil evaluasi menggunakan algoritma K-Nearest Neighbors (KNN), diperoleh nilai akurasi sebesar 0,31. Artinya, model KNN hanya mampu memprediksi dengan benar sekitar 31% dari seluruh data uji yang diberikan. Angka ini menunjukkan bahwa performa model masih tergolong rendah dan belum cukup andal untuk melakukan klasifikasi dengan baik. Hasil ini bisa disebabkan oleh beberapa faktor, seperti pemilihan nilai *k* yang kurang tepat, distribusi data yang tidak merata, atau fitur yang belum dinormalisasi secara optimal. Oleh karena itu, diperlukan langkah-langkah perbaikan seperti tuning parameter, praproses data yang lebih baik, atau mencoba algoritma alternatif yang lebih sesuai dengan karakteristik dataset yang digunakan.
